```python
array_estelionato = np.array(df_estelionato['estelionato'])
array_estelionato
```

```python
media=np.mean(array_estelionato)
print(f"{media:.2f}")
```

```python
#diferenças em relação à média
diferencas=[x - media for x in array_estelionato]
display(diferencas)
```

```python
#elevando as diferenças ao quadrado
quadrados_das_diferencas=[x ** 2 for x in diferencas]
display(quadrados_das_diferencas)
```

```python
# variância, desvio padrão e coeficiente de variação

variancia = df_estelionato['estelionato'].var()
desvio_padrao = df_estelionato['estelionato'].std()
coef_var = desvio_padrao / media_estelionato * 100
print(f'A variância é {variancia}')
print(f'O desvio padrão é {desvio_padrao}')
print(f'O coeficiente de variação é {coef_var}')
```

```python
#calcular a média dos quadrados das diferenças
variancia=sum(quadrados_das_diferencas)/len(quadrados_das_diferencas)
print(variancia)
```

```python
#calcular o desvio-padrão (raiz quadrada da variância)
desviopadrao=variancia**0.5
print(desviopadrao)
```

```python
#coeficiente de variação (CV) é o desvio-padrão dividido pela média
coef_variacao=desviopadrao/media
print(coef_variacao)
```

```python
#cálculo da distância da variância em relação à média
distancia=variancia / (media**2)
print(distancia*100,'%')
```

Como eu interpreto isso?

distância <= 10%: minha dispersão é BAIXA;
distância > 10% e <= 25%: minha dispersão é MODERADA;
distância > 25%: minha dispersão é ALTA.

```python
# assimetria e curtose
assimetria = df_estelionato['estelionato'].skew()
curtose = df_estelionato['estelionato'].kurtosis()
print(f'A assimetria é {assimetria}')
print(f'A curtose é {curtose}')

#interpretação da assimetria

if -0.5 <= assimetria <= 0.5:
    resultado="Distribuição Simétrica"
elif assimetria > 0.5:
    resultado="Distribuição Assimétrica Positiva"
else:
    resultado="Distribuição Assimétrica Negativa"

print(resultado)

#interpretação da curtose

if 2.5 <= curtose <= 3.5:
    resultado2="Distribuição Mesocúrtica"
elif curtose < 2.5:
    resultado2="Distribuição Platicúrtica"
else:
    resultado2="Distribuição Leptocúrtica"

print(resultado2)
```
