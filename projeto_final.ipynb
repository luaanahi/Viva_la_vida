```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BaseDPEvolucaoMensalCisp.csv', sep = ';', encoding = 'latin1')
```

```python
for i in df.columns:
    print(i)
```

```python
display(df.head())
```

```python
df_estelionato = df[['munic', 'estelionato']].copy()
display(df_estelionato)
```

```python
estatistica=df.describe()
display(estatistica)
df_copia=df.copy()
display(df_copia)
```

```python
df_estelionato = df[['munic', 'estelionato']].groupby('munic').sum().reset_index()
df_estelionato
```

```python
array_estelionato = np.array(df_estelionato['estelionato'])
array_estelionato
```

```python
media_estelionato = np.mean(array_estelionato)
media_estelionato
```

```python
mediana_estelionato = np.median(array_estelionato)
mediana_estelionato
```

```python
q1 = np.quantile(array_estelionato, 0.25)
q2 = np.quantile(array_estelionato, 0.50)
q3 = np.quantile(array_estelionato, 0.75)
print(f'O Q1 é {q1}, o Q2 é {q2} e o Q3 é {q3}')
```

```python
distancia=(media_estelionato-mediana_estelionato)/mediana_estelionato*100
print (f'Distância entre média e mediana: {distancia:.2f}%')
```

```python
maximo = np.max(array_estelionato)
minimo = np.min(array_estelionato)
amplitude = maximo - minimo
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)

print(f'Máximo: {maximo}')
print(f'O limite superior é {limite_superior}')
print(f'Mínimo: {minimo}')
print(f'O limite inferior é {limite_inferior}')
print(f'Q1:{q1}')
print(f'Q2:{q2}')
print(f'Q3:{q3}')
print(f'Média:{media_estelionato}')
print(f'Mediana:{mediana_estelionato}')
print(f'O IQR é {iqr}')
```

```python
estelionato_outliers = df_estelionato[df_estelionato['estelionato'] >limite_superior]
estelionato_outliers = estelionato_outliers.sort_values(by = 'estelionato', ascending= False)
estelionato_outliers
```

```python
#Medidas de Dispersão
valor_maximo = df_estelionato['estelionato'].max()
valor_minimo = df_estelionato['estelionato'].min()
amplitude_total = valor_maximo - valor_minimo

print(f'Menor Valor:{valor_minimo}')
print(f'Maior Valor:{valor_maximo}')
print(f'Amplitude:{amplitude_total}')
```

```python
import matplotlib.pyplot as plt
#configuração do painel
fig,ax=plt.subplots(1,2,figsize=(12,6))

#Boxplot
ax[0].boxplot(df_estelionato['estelionato'])
ax[0].set_title('Distribuição de Dados - Boxplot')
ax[0].set_ylabel('Casos de Estelionato')

#Exibindo as Medidas
ax[1].text(0.1,0.8,f"Menor Valor:{valor_minimo}")
ax[1].text(0.1,0.7,f"Maior Valor:{valor_maximo}")
ax[1].text(0.1,0.6,f"Amplitude Total:{amplitude_total}")
ax[1].set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# variância estelionato, devio padrão e coeficiente de variação

variancia = df_estelionato['estelionato'].var()
desvio_padrao = df_estelionato['estelionato'].std()
coef_var = desvio_padrao / media_estelionato * 100
print(f'A variância é {variancia}')
print(f'O desvio padrão é {desvio_padrao}')
print(f'O coeficiente de variação é {coef_var}')
```

```python
# assimetria e curtose
assimetria = df_estelionato['estelionato'].skew()
curtose = df_estelionato['estelionato'].kurtosis()
print(f'A assimetria é {assimetria}')
print(f'A curtose é {curtose}')
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(df_estelionato['estelionato'])
plt.title('Boxplot do Estelionato', fontsize=14)
plt.ylabel('Estelionato', fontsize=12)

plt.ylim(5,1000)
plt.show()
```
