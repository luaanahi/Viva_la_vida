```python
df_estelionato = df[['munic', 'estelionato']].groupby('munic').sum().reset_index()
df_estelionato
```

```python
array_estelionato = np.array(df_estelionato['estelionato'])
array_estelionato
```

```python
media_estelionato = np.mean(array_estelionato)
media_estelionato
```

```python
mediana_estelionato = np.median(array_estelionato)
mediana_estelionato
```

```python
q1 = np.quantile(array_estelionato, 0.25)
q2 = np.quantile(array_estelionato, 0.50)
q3 = np.quantile(array_estelionato, 0.75)
print(f'O Q1 é {q1}, o Q2 é {q2} e o Q3 é {q3}')
```

```python
estelionato_menores = df_estelionato[df_estelionato['estelionato'] <q1]
estelionato_menores = estelionato_menores.sort_values(by = 'estelionato')
estelionato_menores
```

```python
estelionato_maiores = df_estelionato[df_estelionato['estelionato'] > q3]
estelionato_maiores = estelionato_maiores.sort_values(by = 'estelionato', ascending = False)
estelionato_maiores
```

```python
maximo = np.max(array_estelionato)
minimo = np.min(array_estelionato)
amplitude = maximo - minimo
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)

print(f'Mínimo: {minimo}')
print(f'O limite inferior é {limite_inferior}')
print(f'Q1:{q1}')
print(f'Q2:{q2}')
print(f'Q3:{q3}')
print(f'Média:{media_estelionato}')
print(f'Mediana:{mediana_estelionato}')
print(f'O IQR é {iqr}')
print(f'Máximo: {maximo}')
print(f'O limite superior é {limite_superior}')
```
